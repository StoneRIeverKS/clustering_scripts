# import

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import math
```

# 乱数の発生

```python
#乱数の生成
np.random.seed(0)
x1 = np.array([np.random.normal(loc, 1, 20) for loc in [2, 10, -2]]).flatten() #ランダムな80個の数を生成
x2 = np.array([np.random.normal(loc, 1, 20) for loc in [0, 10, 0]]).flatten() #ランダムな80個の数を生成

#入力データ
X = np.c_[x1, x2]
```

```python
#散布図で表示
plt.rcParams["font.family"] = "Hiragino Kaku Gothic Pro"
plt.scatter(x1, x2)
plt.xlim(-10, 20)
plt.ylim(-3, 15)
plt.legend()
plt.grid()
plt.show()
```

# ユークリッド距離でk-meansを実行

```python
#エルボー法を用いてクラスタ数を決定する
distortions = []

for i  in range(1,11):                # 1~10クラスタまで一気に計算 
    km_model_euclid = KMeans(
        n_clusters=i,
        init='k-means++',     # k-means++法によりクラスタ中心を選択
        n_init=10,
        max_iter=300,
        random_state=0)
    km_model_euclid.fit(X)                         # クラスタリングの計算を実行
    distortions.append(km_model_euclid.inertia_)   # km.fitするとkm.inertia_が得られる

plt.plot(range(1,11),distortions,marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

```python
#kmeansを実装
km_model_euclid = KMeans(
    n_clusters = 2,
    init = "random",
    n_init = 10,
    max_iter = 300,
    tol = 1e-04,
    random_state = 1,
    )

#モデルへのあてはめ
y_km_euclid = km_model_euclid.fit(X)

#結果をプロット
plt.rcParams["font.family"] = "Hiragino Kaku Gothic Pro"
plt.scatter(x1, x2, c = y_km_euclid.labels_, s = 30)
plt.xlim(-10, 20)
plt.ylim(-3, 15)
plt.title("k-means_euclid")
plt.legend()
plt.grid()
plt.show()
```
