# import

```python
import numpy as np
from math import atan2, sin
import matplotlib.pyplot as plt
from itertools import combinations
import pandas as pd

#散布図の表示などに用いるライブラリ
import matplotlib.colors as plc 
import matplotlib.pyplot as plt 
from IPython.display import display, HTML 
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline

from sklearn.cluster import KMeans
```

# 極座標⇒直行座標のメソッド

```python
def get_pc(cartesian_coordinate_vec):
    
    cartesian_coordinate_vec = np.array(cartesian_coordinate_vec) #NumPy arrayへの変換
    dim = len(cartesian_coordinate_vec) #入力ベクトルの次元数を取得
    r = np.linalg.norm(cartesian_coordinate_vec) #入力ベクトルのノルムを取得
        
    if dim == 0:
        print("2次元以上のベクトルを入力してください")
    elif r == 0:
        return(np.zeros(shape=dim)) #入力ベクトルが0の場合、極座標も全て0を返す
    elif dim == 1:
        print("2次元以上のベクトルを入力してください") #2次元以上のベクトルしか極座標変換できない
    elif dim == 2:
        polar_coordinate_vec = np.array([]) #出力ベクトルを入れるベクトルを用意
        
        polar_coordinate_vec = np.append(polar_coordinate_vec, r) #出力ベクトルにノルムを格納
        polar_coordinate_vec = np.append(polar_coordinate_vec, atan2(cartesian_coordinate_vec[1], cartesian_coordinate_vec[0])) 
        
        return(polar_coordinate_vec)
    else:
        polar_coordinate_vec = np.array([]) #出力ベクトルを入れるベクトルを用意
        polar_coordinate_vec = np.append(polar_coordinate_vec, atan2(cartesian_coordinate_vec[1], cartesian_coordinate_vec[0])) 
        
        for i in range(2, len(cartesian_coordinate_vec)):
            tmp = atan2(cartesian_coordinate_vec[i]*sin(polar_coordinate_vec[i-2]), cartesian_coordinate_vec[i-1])
            polar_coordinate_vec = np.append(polar_coordinate_vec, tmp)
        
        polar_coordinate_vec = np.append(polar_coordinate_vec, r)
        polar_coordinate_vec = polar_coordinate_vec[::-1]
        
        return(polar_coordinate_vec)
```

# シミュレーションデータ作成

```python
#乱数の生成
np.random.seed(0)
x1 = np.array([np.random.normal(loc, 0.5, 20) for loc in [1, -1, 10]]).flatten() #ランダムな60個の数を生成
x2 = np.array([np.random.normal(loc, 0.5, 20) for loc in [0, 0, 0]]).flatten() #ランダムな60個の数を生成
x3 = np.array([np.random.normal(loc, 0.5, 20) for loc in [0, 0, 0]]).flatten() #ランダムな60個の数を生成

#入力データ
X = np.c_[x1, x2, x3]

#pandas.DataFrameに変換
X_df = pd.DataFrame(X)
```

```python
#図形の箱用意
fig = plt.figure()
ax1 = Axes3D(fig)
 
# ラベル設定
ax1.set_xlabel('x1')
ax1.set_ylabel('x2')
ax1.set_zlabel('x3')
 
# 散布図作成
colorlist = ["r"]*20+["b"]*20+["y"]*20
ax1.scatter3D(x1, x2, x3, color = colorlist)
plt.legend()
 
# 表示
plt.show()
plt.close()
```

```python
#散布図で表示
plt.scatter(x1, x2, c=colorlist)
plt.xlim(-10, 20)
plt.ylim(-3, 15)
plt.legend()
plt.grid()
plt.show()
```

# 直行座標⇒極座標の実行

```python
#直交座標から極座標への変換
X_pc = X_df.apply(get_pc, axis=1)

#角度だけ抽出
X_radians = pd.DataFrame()
for x in X_pc:
    X_radians = X_radians.append(pd.Series(x[1:]), ignore_index=True)
```

# k-meansの実行

```python
#kmeansを実装
km_pc = KMeans(
    n_clusters = 2,
    random_state = 10,
    init = "random"
    )

#モデルへのあてはめ
y_km_pc = km_pc.fit(X_radians)

#結果をプロット
plt.scatter(x1, x2, c = y_km_pc.labels_, s = 30)
plt.xlim(-10, 20)
plt.ylim(-3, 15)
plt.title("k-means_pc")
plt.legend()
plt.grid()
plt.show()
```

# 結論
極座標変換した場合は、角度の範囲が-π～πである。このとき、-πとπ付近の点はその間の角が小さいにもかかわらず、極座標においては最も離れた点として扱われてしまう。したがって、0～πと-π～0でクラスタ分割されてしまうため、コサイン類似度とは異なる結果となってしまう。
